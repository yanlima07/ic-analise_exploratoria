```python
x = ('GreeNC Phytozome','CANTATAdb 2.0','PLncDB','PLncDB RNAcentral','PLncDB Liu','PLncDB EVL')
y = [len(seqAthalianaGreeNCphytozome), len(seqAthalianaCANTATA), len(seqAthalianaPLncDB), len(seqAthalianaPLncDBrnacentral), len(seqAthalianaPLncDBliu), len(seqAthalianaPLncDBevl)]

fig, (ax, axZ, axO) = plt.subplots(1,3)

ax.bar(x, y)
ax.set_title("Número de sequências lncRNA (A. thaliana)")

for x1, y1 in zip(x, y):
  ax.annotate(y1, (x1, (y1 + 100)))


xZ = ('GreeNC Phytozome', 'GreeNC Ensembl', 'CANTATAdb 2.0', 'PLncDB','PLncDB RNAcentral','PLncDB Li','PLncDB EVL')
yZ = [len(seqZmaysGreeNCphytozome), len(seqZmaysGreeNCensembl), len(seqZmaysaCANTATA), len(seqZmaysPLncDB), len(seqZmaysPLncDBrnacentral), len(seqZmaysPLncDBli), len(seqZmaysPLncDBevl)]

axZ.bar(xZ, yZ)

axZ.set_title("Número de sequências lncRNA (Z. mays)")

for x1, y1 in zip(xZ, yZ):
  axZ.annotate(y1, (x1, (y1 + 100)))

xO = ('GreeNC Phytozome','CANTATAdb 2.0','PLncDB','PLncDB RNAcentral','PLncDB Cameron','PLncDB EVL')
yO = [len(seqOsatGreeNCphytozome), len(seqOsatCANTATA), len(seqOsatPLncDB), len(seqOsatPLncDBrnacentral), len(seqOsatPLncDBcameron), len(seqOsatPLncDBevl)]

axO.bar(xO, yO)
axO.set_title("Número de sequências lncRNA (O. sativa)")

for x1, y1 in zip(xO, yO):
  axO.annotate(y1, (x1, (y1 + 100)))

fig.set_size_inches(35, 8)
plt.style.use('seaborn')
plt.show()
```
